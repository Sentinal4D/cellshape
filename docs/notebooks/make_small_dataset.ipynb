```python
import pandas as pd

full_df = pd.read_csv(
    "/home/mvries/Documents/Datasets/OPM/SingleCellFromNathan_17122021/all_cell_data.csv"
)

full_df["Treatment"].value_counts()
```

```python
import numpy as np

size = 100  # sample size
replace = False  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace), :]
small_df = (
    full_df.groupby("Treatment", as_index=False)
    .apply(fn)
    .reset_index()
    .iloc[:, 3:]
)
small_df
```

```python
small_df.to_csv(
    "/home/mvries/Documents/CellShape/UploadData/FullDataset/small_data.csv"
)
```

```python
import shutil
import os

img_dir = "/home/mvries/Documents/CellShape/UploadData/FullDataset/"
new_img_dir = "/home/mvries/Documents/CellShape/UploadData/SmallDataset/"
cell_component = "cell"
for idx in range(len(small_df)):
    treatment = small_df.loc[idx, "Treatment"]
    plate_num = "Plate" + str(small_df.loc[idx, "PlateNumber"])

    if cell_component == "cell":
        component_path = "stacked_pointcloud"
    else:
        component_path = "stacked_pointcloud_nucleus"

    img_path = (
        os.path.join(
            img_dir,
            plate_num,
            component_path,
            treatment,
            small_df.loc[idx, "serialNumber"],
        )
        + ".ply"
    )

    new_path = (
        os.path.join(
            new_img_dir,
            plate_num,
            component_path,
            treatment,
            small_df.loc[idx, "serialNumber"],
        )
        + ".ply"
    )
    os.makedirs(
        os.path.join(new_img_dir, plate_num, component_path, treatment),
        exist_ok=True,
    )

    shutil.copy(img_path, new_path)
```
